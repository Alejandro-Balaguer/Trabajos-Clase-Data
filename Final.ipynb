```python
import numpy as np

# Funciones de colocación de barcos

def puede_colocar(tablero, fila, col, eslora, orientacion):
    if orientacion == "H":
        fila0 = max(0, fila - 1)
        fila1 = min(9, fila + 1)
        col0 = max(0, col - 1)
        col1 = min(9, col + eslora)
    else:  # "V"
        fila0 = max(0, fila - 1)
        fila1 = min(9, fila + eslora)
        col0 = max(0, col - 1)
        col1 = min(9, col + 1)
    return not np.any(tablero[fila0:fila1+1, col0:col1+1] == "B")


def colocar_barco(tablero, eslora):
    while True:
        orientacion = np.random.choice(["H", "V"])
        if orientacion == "H":
            fila = np.random.randint(0, 10)
            col  = np.random.randint(0, 10 - (eslora - 1))
        else:
            fila = np.random.randint(0, 10 - (eslora - 1))
            col  = np.random.randint(0, 10)
        if puede_colocar(tablero, fila, col, eslora, orientacion):
            if orientacion == "H":
                tablero[fila, col:col+eslora] = "B"
            else:
                tablero[fila: fila+eslora, col] = "B"
            return

# Función de disparo interactiva

def dispara(tablero, fila, col):
    if tablero[fila, col] == "B":
        tablero[fila, col] = "X"
        print(f"Disparo en ({fila},{col}): Un Ingles menos!!")
       
    elif tablero[fila, col] == "X":
        print(f"Disparo en ({fila},{col}): Chico, ese ingles ya esta muerto!!")
        
    else:
        tablero[fila, col] = "-"
        print(f"Disparo en ({fila},{col}): Cada vez que fallas tu madre llora")
        

# Inicializar tablero y colocar flota

tablero = np.full((10, 10), "~")
esloras = [1,1,1,1,2,2,2,3,3,4]
for e in esloras:
    colocar_barco(tablero, e)

# Bucle interactivo de disparos

print("¡Comienza el juego de Hundir la Flota Inglesa!")
print("Introduce coordenadas entre 0 y 9. Para salir, escribe 'salir'.")

while True:
    comando = input("Fila,Columna > ")
    if comando.lower() == 'salir':
        print("Juego terminado por el usuario.")
        break
    try:
        fila_str, col_str = comando.split(',')
        fila = int(fila_str)
        col = int(col_str)
        if not (0 <= fila < 10 and 0 <= col < 10):
            raise ValueError
    except ValueError:
        print("Entrada no válida. Formato: fila,col (ej. '3,5')")
        continue

    resultado = dispara(tablero, fila, col)

    # Mostrar estado parcial del tablero
    for f in tablero:
        print(" ".join(f))

    # Comprobar fin de juego
    if not np.any(tablero == 'B'):
        print("\nEnhorabuena! Has hundido toda la flota inglesa. Un gran dia para el Imperio.")
        break
```
